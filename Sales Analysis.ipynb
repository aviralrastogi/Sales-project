```python
# Q what products most often sold together ?  #important

        
        
```

```python
data.head(5)
```

```python
data.rename(columns={"Order ID":"Orderid"},inplace=True)
```

```python
data.head(10)
```

```python
df=data[data["Orderid"].duplicated(keep=False)]
df["Grouped"]=df.groupby("Orderid")["Product"].transform(lambda x:','.join(x))
df=df[["Orderid","Grouped"]].drop_duplicates()
df.head(10)
```

```python
from itertools import combinations
from collections import Counter
count=Counter()
for row in df["Grouped"]:
    row_list=row.split(',')
    count.update(Counter(combinations(row_list,2)))  # This line counts how many times this line has been occured 
count.most_common(10)
```
